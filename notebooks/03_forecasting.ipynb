# 03: Forecasting and Evaluation

```
import sys, os
sys.path.append(os.path.abspath('..'))

from src.preprocessing import load_data, clean_data
from src.feature_engineering import split_data, get_preprocessor
from src.train import build_pipeline, train_model
from src.evaluate import print_evaluation, plot_precision_recall_curve, get_feature_importances
```

```
df = load_data("../data/raw/WA_Fn-UseC_-Telco-Customer-Churn.csv")
df = clean_data(df)
X_train, X_test, y_train, y_test = split_data(df)
preprocessor, num_cols, cat_cols = get_preprocessor()
```

```
log_pipeline = build_pipeline(preprocessor, model_type="logistic")
log_model = train_model(log_pipeline, X_train, y_train)
```

```
print("--- Logistic Regression Evaluation ---")
print_evaluation(log_model, X_test, y_test)
```

```
plot_precision_recall_curve(log_model, X_test, y_test)
```

```
print("--- Logistic Regression with custom threshold (0.35) ---")
print_evaluation(log_model, X_test, y_test, threshold=0.35)
```

```
rf_pipeline = build_pipeline(preprocessor, model_type="random_forest")
rf_model = train_model(rf_pipeline, X_train, y_train)
```

```
print("--- Random Forest Evaluation ---")
print_evaluation(rf_model, X_test, y_test)
```

```
feature_importances = get_feature_importances(rf_model, num_cols, cat_cols)
display(feature_importances.head(10))
```
